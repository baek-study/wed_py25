## **1.1 인수 유형**

*   위치 인수 : 인수를 순서대로 값을 전달
* 기본값(디폴트) 인수  : 매개변수에 기본값 정의
*   키워드 인수 : 이름  - 값
* 가변 인수 : 인수 개수가 정해지지 않음

```python
## 예제 - 위치 인수
# 함수 정의
# 함수 이름 : get_area
# 매개변수 : radius,
# 반환값 : 결과(return) 있음 - area
# 하는 일 : area에 radius*radius*3.14 를 계산
def get_area(radius) :
    area = radius  * radius * 3.14
    return area

# 함수 호출
value = get_area(5)
print(f'{value = }')
```

```python
## 예제 - 기본값 인수
# 함수 정의
# 함수 이름 : sum_para
# 매개변수 : v1, v2, v3
# 반환값 : 결과(return) 있음 - result
# 하는 일 : 매개변수를 합하여 result에 넣음
def sum_para(v1, v2, v3=0) :
    print(f'{v1 = }, {v2 = }, {v3 = }')
    result = v1 + v2 + v3
    return result

# 함수 호출
value = sum_para(1, 2, 3)
print(f'{value = }')

value = sum_para(1, 2)
print(f'{value = }')
```

```python
## 예제 - 키워드 인수
# 함수 정의
# 함수 이름 : sub
# 매개변수 : x, y, z
# 반환값 : 결과(return) 없음
# 하는 일 : x, y, z 값을 프린트

## sub 함수 호출
def sub(x, y, z) :
    print(f'{x = },{y = }, {z = }')

## print 함수 호출
sub(2,3,4)
sub(3, y=4, z=5)
sub(3, z=5, y=4)
sub(z = 5, y = 4, x = 3)
sub(2, z=3, y=4)

## sum_para 함수 호출
value = sum_para(v1 = 3, v3= 5, v2 =4)

## print 함수
print("hello", end = "...")
print("world")
```

```python
## 예제 - 가변 인수
## print 함수 호출
print("hello", "world")
print(1, 2, 3, 4, 5)

# 함수 정의
# 함수 이름 : print_var
# 매개변수 : *x
# 반환값 : 결과(return) 없음
# 하는 일: 매개변수 프린트
def print_var( *x ) :
    print(x)

# print_var 호출
print_var(1)
print_var(1, 2)
print_var(1,2,3,4,5)
print_var("hello", "world")
```

##  **1.2 함수와 리스트**


### **가. 매개변수가 리스트**
- 인수로 리스트 지정 시 리스트 객체에 대한 주소가 매개변수로 전달
- 함수 안에서 매개변수 변경 시 원본도 바뀜

```python
# 예제 - 매개변수가 리스트
# 함수 정의
# 함수 이름 : func1
# 매개변수 : para_list
# 반환값 : 결과(return) 없음
# 하는 일: para_list의 주소 출력, para_list 출력
def func1(para_list) :
    print(f'{para_list=}')
    print(f'{id(para_list)=}\n')
    para_list[0] = 99
    print(f'함수내 - para_list[0] = 99 ==>\n{para_list=}\n')

# 리스트 생성 nums에 0,1,1,2,3,5,8 넣기
nums = [0,1,1,2,3,5,8]

# nums 값/주소 출력
print(f'{nums=}')
print(f'{id(nums)=}\n')

# func1 함수 호출 - 이때 인수는 리스트
func1(nums)

print(f'함수 호출후 : {nums=}')
print(f'함수 호출후 : {id(nums)=}')
```

```python
# 예제 - 매개변수가 리스트 - 내부에서 새로운 리스트로 생성 예제
# 함수 정의
# 함수 이름 : func1
# 매개변수 : para_list
# 반환값 : 결과(return) 없음
# 하는 일: para_list의 주소 출력, para_list 출력
def func1(para_list) :
    print(f'{para_list=}')
    print(f'{id(para_list)=}\n')
    para_list = [1,2,3,4]
    print(f'para_list = [1,2,3,4] ==>\n{para_list=}')
    print(f'{id(para_list)=}\n')

# 리스트 생성 nums에 0,1,1,2,3,5,8 넣기
nums = [0,1,1,2,3,5,8]

# nums 값/주소 출력
print(f'{nums=}')
print(f'{id(nums)=}\n')

# func1 함수 호출 - 이때 인수는 리스트
func1(nums)

print(f'함수 호출후 : {nums=}')
```

### **나. 반환값이 리스트**
- 리스트 객체에 대한 아이디를 호출 측으로 반환

```python
# 예제 - 반환값이 리스트
# 함수 정의
# 함수 이름 : create_list
# 매개변수 : 없음
# 반환값 : 리스트(inner_list) 를 반환
# 하는 일: inner_list에 1,2,3,4 생성한 후,  주소와 리스트 출력후 반환
def create_list() :
    inner_list = [1,2,3,4]
    print(f'{inner_list=}')
    print(f'{id(inner_list)=}\n')
    return inner_list

# 함수 호출
# 호출 한 후 result에 대입
value = create_list()

# 주소와 내용 출력
print(f'{value=}')
print(f'{id(value)=}')
```

# **2부 지역 변수 vs 전역변수**



## **2.1 지역변수**
*  함수 내에서 생성된 변수와 매개변수
  - 선언된 함수 내에서만 접근 가능
  - 지역 변수 마다 동일한 이름 사용 가능 - 서로 다른 영역



```python
# 예제1 : 지역변수 -  선언된 함수 내에서만 접근 가능
# 함수 정의 func1  :  지역변수 lx에 10 넣고 출력

def func1() :
    lx = 10
    print(f'{lx=}')
    print(f'{id(lx)=}')

# 주 프로그램 - 함수 호출 func1
func1()

# lx를 접근 - 오류
print(f'{lx=}')
```

```python
# 예제2 : 지역변수 -  다른 함수에서 동일한 함수 이름 가능 - 서로 다른 영역
# 함수 정의 func2  :  지역변수 lx에 20 넣고 출력
def func2 () :
    lx = 20
    print(f'{lx=}')
    print(f'{id(lx)=}')

# 주 프로그램 - 함수 호출 func1, func2
func1()
func2()

# lx를 접근 - 오류
print(f'{lx=}')
```

## **2.2 전역 변수**
- 함수 외부에서 생성되며 모든 함수에서 접근 가능

- 함수 내부에서 전역변수 변경시 :  global 키워드 이용

```python
# 전역변수 - 함수 외부에 생성 되어 모두 접근, gx = 10
gx = 10

# 함수 정의 func3 : 전역 변수 출력
def func3() :
    print(f'func3  : {gx=}')

# 함수 정의 func4
def func4() :
    gx = 20        # 지역 변수로 생성 - 전역 변수와 이름
    print(f'func4  : {gx=}')


# 주 프로그램 - 함수 호출 func3, func4
func3()
func4()
print(f'함수밖 : {gx=}')
```

```python
# 전역변수 - 전역 변수 변경시 global 키워드 이용
gx = 10

# 함수 정의 func3 : 전역 변수 출력
def func3() :
    print(f'{gx=}')
    print(f'{id(gx)=}')

def func5() :
    global gx
    gx = 20
    print(f'func5 : {gx=}')
    print(f'func5 : {id(gx)=}\n')

# 주 프로그램 - 함수 호출 func3, func4
func3()
func5()
print(f'함수 밖 : {gx=}')
print(f'함수 밖 : {id(gx)=}')
```

## 친구 연락처 관리 프로그램

```python
## manage 모듈 - 친구 연락처 관리 모듈
%%writefile manage.py

## 연락처 출력 함수 : print_contacts -  매개변수 : contacts
def print_contacts(contacts):
    print("연락처 출력 함수 실행")
    for contact_list in contacts :
        print(contact_list)
        print(f'이름:{contact_list[0]} 전화번호 : {contact_list[1]}')

## 연락처 추가 함수 : add_contacts - 매개변수 : contacts
def add_contacts(contacts) :
    print("연락처 추가 함수 실행")

    name = input("이름 입력 : ")
    tel = input("전화번호 입력 : ")
    contact_list  = [name, tel]
    contacts.append(contact_list)

## 연락처 삭제 함수 : delete_contacts - 매개변수 : name, contacts
def delete_contacts(delete_name ,contacts) :
    print("연락처 삭제 함수 실행")

    found = False

    for contact_list in contacts :
        if contact_list[0] == delete_name :
            found = True
            delete_contact = contact_list
            break
    if found :
        contacts.remove(delete_contact)

## 전화번호 검색 함수 : search_contacts - 매개변수 : name, contacts, 반환 : 전화번호
def search_contacts(search_name, contacts) :
    print("연락처 검색 함수 실행")
    found = False

    for contact_list in contacts :
        if contact_list[0] == search_name :
            found = True
            tel_number = contact_list[1]    ### 전화 번호 (삭제와 다른 내용)
            break

    if found :
        return tel_number
    else :
        return "연락처에 없음"
```

```python
## 모듈 사용
import manage

# 예제 - 친구 연락처 관리

contacts = []  # 연락처 관리하는 2 차원 리스트

while True :

    # 메뉴 출력
    print("--------------------")
    print("1. 친구 연락처 출력")
    print("2. 친구 연락처 추가")
    print("3. 친구 연락처 삭제")
    print("4. 전화번호 검색")
    print("9. 종료")
    print("--------------------")

    menu = int(input("메뉴를 입력하세요 : "))
    if menu == 1:
        manage.print_contacts(contacts)
    elif menu ==2:
        manage.add_contacts(contacts)
    elif menu == 3:
        delete_name = input("삭제할 이름 입력 : ")
        manage.delete_contacts(delete_name,contacts)

    elif menu == 4:
        search_name = input("검색할 이름 입력 : ")
        value = manage.search_contacts(search_name,contacts)
        print(f'{search_name} 연락처 : {value} ')
    elif menu == 9:
        print("종료합니다.")
        break
```
