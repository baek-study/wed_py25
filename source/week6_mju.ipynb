# **1부 논리 연산자**



## **1.1 논리연산자**
- 여러 개의 조건을 조합하여 참과 거짓 <br>
- and, or, not

```python
# 1.1 논리 연산자 연습
# n1에는 참을 n2에는 거짓을 대입
n1 = True
n2 = False

# n1과 n2의 and 출력
print(f'{n1 and n2 = }')

# n1과 n2의 or 출력
print(f'{n1 or n2 = }')

# n1의 not 출력
print(f'{not n1 = }')

# n2의 not 출력
print(f'{not n2 = }')
```

## **1.2 예제1 - 값의 범위 표현**

num이 10보다 크고, num이 20보다 작다 표현

```python
# 1.2 논리 연산자 실습 예제 1
num = 5
# num이 10 보다 크고, 20 보다 작다
# 1. 수학적 표현 - 파이썬 가능
print(f'{10 < num < 20 = }')

# 2. and 이용한 표현
print(f'{10 < num  and num < 20 =}')

# 3. 변수가 왼쪽에 있는 표현 - 가독성 높인 경우
print(f'{num > 10 and num < 20 = }')
```

## **1.2 예제2**
age가 유효한 나이인지 (0~120사이)

```python
# 1.2 논리 연산자 실습 예제 2
# age를 사용자 입력 - 메시지 : 당신의 나이는  - 정수
age = int(input('당신의 나이는? : '))

# age가 0보다 크거나 같고 120보다 작거나 같음 - 유효한 나이

# res1 : 수학적 표현
res1 = 0 <= age <= 120
# res2 : and 이용한 표현
res2 = 0 <= age and age <= 120
# res3 : 변수가 왼쪽에
res3 = age >= 0 and age <= 120

# 유효한 나이인지.. 출력 - res1, res2, res3
print(f'{age}는 유효한 나이인가? {res1} {res2} {res3}')

# age가 0보다 작거나 120 보다 큼 - 유효한 나이가 아님 - res4
res4 = age < 0 or age > 120

# 유효한 나이가 아닌지 출력 - res4
print(f'{age}는 유효한 나이가 아닌가? {res4}')
# 유효한 나이인지 - not res4
print(f'{age}는 유효한 나이인가? {not res4}')

```

## **1.3 연산자 우선 순위**
- 산술>관계>논리>할당

```python
# 1.3 연산자 우선 순위 연습
age = 10

result = age-5 >= 0 and age <=12
print(f'{result=}')

result = (age-5 >= 0) and (age <=12)
print(f'{result=}')
```

## **3.2 if구문 예제1**


1. 한줄 버전 : if 조건: 문장
2. 표준 버전<br>
if 조건: <br>
&ensp;&ensp;&ensp;&ensp;문장
3. 표준 확장 - 코드 블록<br>
if 조건: <br>
&ensp;&ensp;&ensp;&ensp;문장1<br>
&ensp;&ensp;&ensp;&ensp;문장2



```python
# 3.2 if 조건문 예제 1
# 사용자가 입력한 2개의 정수n1, n2의 비교
# n1 - 첫번째 정수
# n2 - 두번째 정수
n1 = int(input('첫번째 정수 : '))
n2 = int(input('두번째 정수 : '))

# 관계연산자
# n1이 n2 보다 큰지
n1 > n2

# 만약 n1이 n2 보다 큰지? - 버전 1 - 한줄에
if n1 > n2 : print(f'{n1}이 {n2}보다 크다')

# n1이 n2 보다 큰지? - 버전 2  - 기본 버전
if n1 > n2 :
    print(f'{n1}이 {n2}보다 크다')

# n1이 n2 보다 큰지? - 버전 3  - 여러 문장 제어 받음 = n1이 n2 보다 크다, n1이 더 크다
if n1 > n2 :
    print(f'{n1}이 {n2}보다 크다')
    print(f'{n1}이 크다')

print('끝')
```

## **3.2 if구문 예제2**

*   만약 n1이 n2 보다 크면?
*   만약 n1이 n2 보다 작거나 같으면?




```python
# 3.2 if 조건문 예제 2
# 사용자가 입력한 2개의 정수n1, n2의 비교
# n1 - 첫번째 정수
# n2 - 두번째 정수
n1 = int(input('첫번째 정수 : '))
n2 = int(input('두번째 정수 : '))

# 배타적인 상황
# 만약 n1이 n2 보다 크면? n1이 n2보다 더 크다 출력
if n1 > n2 :
    print(f'{n1}이 {n2}보다 크다')

# 만약 n1이 n2 보다 작거나 같으면? n1이 n2보다 작거나 같다 출력
if n1 <= n2 :
    print(f'{n1}이 {n2}보다 작거나 같다')
```

```python
# One more!! - 배타적인 상황
score = 80
# 만약 score가 60 이상(크거나 같으면) "pass"
if score >= 60 :
    print("pass")

# 만약 score가 60 보다 작으면 "fail"
if score < 60 :
    print("fail")
```

*   **[문제3]** 사용자가 입력 받은 연도가 윤년인지 판별하는 프로그램
- 조건A: 4로 나누어 떨어지지만, 100으로는 나누어 떨어지지 않는해
- 조건B: 400으로 나누어 떨어지는해
- image.png

---
연도 : 2025(사용자 입력) <br>
윤년 아님<br>
프로그램 종료.

```python
# 실습 문제 3
# 연도 입력
year = int(input('연도 : '))

# 조건A: 4로 나누어 떨어지지만, 100으로는 나누어 떨어지지 않는해
cond_a = (year % 4 == 0)  and  (year % 100 != 0)

# 조건B: 400으로 나누어 떨어지는해
cond_b = (year % 400 == 0)

# 조건A 또는 조건B 이면  "윤년" 을 출력
if cond_a or cond_b :
    print("윤년")

# 그렇지 않으면 "윤년 아님"을 출력
else :
    print("윤년이 아님")
```

```python
# 실습 문제 3
# 연도 입력
year = int(input('연도 : '))

# 조건A: 4로 나누어 떨어지지만, 100으로는 나누어 떨어지지 않는해
isDivisibleBy4Not100 = (year % 4 == 0)  and  (year % 100 != 0)

# 조건B: 400으로 나누어 떨어지는해
isDivisibleBy400 = (year % 400 == 0)

isLeapYear = isDivisibleBy4Not100 or isDivisibleBy400

# 조건A 또는 조건B 이면  "윤년" 을 출력
if isLeapYear :
    print("윤년")

# 그렇지 않으면 "윤년 아님"을 출력
else :
    print("윤년이 아님")
```
