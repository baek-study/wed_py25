# **1부 파일 기본**

## **1.1 파일 열고 닫기 (고전적 방법)**

*  3단계 : 열기(open) - 읽고/쓰기 - 닫기(close)
* 파일모드 : r(읽기), w(쓰기), a(추가)
* 파일열기 : 변수 = open(파일이름, 파일모드)
* 파일닫기 : 변수.close()



```python
#예제1 - 파일 쓰기 모드로 열기 - 닫기 (고전적인 방법)
# exam.txt를 쓰기 모드(w) 로 open 한 후, 변수 file에 넣음
file = open('exam.txt', 'w')

# 닫기
file.close()
print('파일 쓰기를 종료 하였습니다')
```

```python
#예제2 - 파일 쓰기 모드로 열기 - 닫기 (고전적인 방법)
# exam.txt를 읽기 모드(r) 로 open 한 후, 변수 file에 넣음
file = open('exam.txt', 'r')

# 닫기
file.close()
print('파일 읽기를 종료 하였습니다')
```

# **2부 파일 읽기/파일 쓰기**

## **2.1 통째로(한꺼번에) 파일 읽고 쓰기**
 - 파일변수.read() : 한꺼번에 읽기
 - 파일변수.write(문자열) : 문자열 쓰기
    - 문자열 끝에 \n이 들어가야 한줄씩 쓰여짐

```python
#예제5 - 파일변수.write(문자열) 연습

#  with open(...) as 변수 + 인코딩(encoding = 'utf-8' )
with open('exam.txt', 'w', encoding = 'utf-8') as file:

    # "hong\n" 파일 쓰기
    file.write('hong\n')       # 한줄 띄기(\n) 필수
    # "kim\n" 파일 쓰기
    file.write('kim\n')
    file.write('mju\nmaru\n')
    file.write(str(35)+'\n')    # 숫자는 문자열로 변환
    file.write('alice\n')

print(f'파일쓰기가 완료 되었습니다.')
```

```python
#예제6 - 변수 = 파일변수.read() 연습

#  with open(...) as 변수 + 인코딩(encoding = 'utf-8' )
with open('exam.txt', 'r', encoding = 'utf-8') as file:

    # 파일 읽어서 data에 넣기
    data = file.read()         # read 결과는 문자열
    # data 쓰기
    print(data)
    print(type(data))

print(f'파일읽기가 완료 되었습니다.')
```

## **2.2 한 줄씩 파일 읽기**
- for 문 : for 변수 in 파일변수
- [참고] readline( ) 함수 이용

```python
#예제6 - 변수 = 파일변수.read() 연습

#  with open(...) as 변수 + 인코딩(encoding = 'utf-8' )
with open('exam.txt', 'r', encoding = 'utf-8') as file :

    # for 변수 in 파일변수 :
    for line in file:
        print(line)
```

```python
#심화 - readline() - 한줄 읽기 연습

#  with open(...) as 변수 + 인코딩(encoding = 'utf-8' )
with open('exam.txt', 'r', encoding = 'utf-8') as file :
    while True :
        # readline()으로 한 줄 읽고 line 변수에 넣기
        line = file.readline()

        if line == '' : # EOF(end of file)은 ''
            break
        print(line)

print('파일읽기가 완료 되었습니다')
```

## **2.3 리스트와 결합**

* readlines() : 문자열 리스트로 읽음, 통째로 처리
* writelines() : 문자열 리스트를 쓰기, 통째로 처리


```python
#예제7 - 변수 = 파일변수.readlines() 연습 - 문자열 리스트로 읽음, 한꺼번에 읽음

#  with open(...) as 변수 + 인코딩(encoding = 'utf-8' )
with open('exam.txt', 'r', encoding = 'utf-8') as file :
    # 읽은 후 lines 에 저장
    lines = file.readlines()
    # lines 출력 - f 문자열 출력
    print(f'{lines=}')

print('파일읽기가 완료 되었습니다')
```

```python
#예제8 - 변수 = 파일변수.writelines() 연습 - 문자열 리스트를 쓰기, 한꺼번에 쓰기

#  with open(...) as 변수 + 인코딩(encoding = 'utf-8' )
with open('exam.txt', 'w', encoding = 'utf-8') as file :
    lines = ["hong\n", "kim\n", "mju\n"]
    # lines(문자열 리스트) 를 파일에 쓰기
    file.writelines(lines)

print(f'파일쓰기가 완료 되었습니다.')
```

```python
#예제9 - writelines() 연습 - 리스트의 요소가 문자열이 아닌 경우

nums = [1, 2, 3, 4]

# nums 내부의 요소를 문자열로 전환
str_nums = [ ]
for x in nums:
        str_num = str(x) + '\n'
        str_nums.append(str_num)
print(f'{str_nums=}')

# 리스트 함축
str_nums2 = [ str(x) + '\n' for x in nums ]
print(f'{str_nums2=}')

with open('test.txt', 'w', encoding='utf=8') as file :
    file.writelines(str_nums)

print(f'파일쓰기가 완료 되었습니다.')
```

## **3.1 데이터 정제**
- 문자열.strip() : 공백문자(빈칸, \n 등) 없애기
- 문자열.split(구분자) : 구분자로 문자열 나누기

```python
#예제10 - 문자열.strip() - 공백문자(빈칸, \n 등) 없애기

#  with open(...) as 변수 + 인코딩(encoding = 'utf-8' )
with open('exam.txt', 'r', encoding = 'utf-8') as file :
    # for 변수 in 파일변수 :
    for line in file :
        # print(line)
        # 변수 strip 적용
        clean_line = line.strip()
        # 변수 출력
        print(clean_line)
```

```python
#예제11을 위한 파일 생성 - student.csv
data = '철수, 90\n 영희, 80\n 민수, 75\n'

with open('student.csv', 'w', encoding='utf=8') as file :
    file.write(data)

print(f'파일쓰기가 완료 되었습니다.')
```

```python
#예제11 - 문자열.split(구분자) : 구분자로 문자열 나누기

with open('student.csv', 'r', encoding='utf=8') as file :

    for line in file :
        # 공백문자 없애기
        clean_line = line.strip()
        # 구분자를 , 로 하고, name, score 변수에 값 넣기
        name, score = clean_line.split(',')
        # line과 name, score 출력
        print(name, score)
```

## **3.2 예외(except) 처리**
- 예외 : 프로그램이 갑자기 죽음
    - 파일없음 : FileNotFound
    - 자료형변환 오류
- 예외 처리 : try ~ except ~

```python
# 예제12 - 예외 발생 및 예외 처리 연습1

# 파일 없음 에러 수정
try :
    with open('mju.txt', 'r', encoding='utf=8') as file :
        file.read()
except FileNotFoundError :
    print("파일을 찾을수 없습니다")
```

```python
# 예제13 - 예외 발생 및 예외 처리 연습2

# 문자열을 정수로 변환 에러 수정
try :
    with open('exam.txt', 'r', encoding='utf=8') as file :
        for line in file :
            num = int(line)
except FileNotFoundError :
    print("파일을 찾을수 없습니다")
except ValueError:
    print("문자열을 정수로 변환 오류가 있습니다")
except :
    print("오류")
```

```python
## manage_contacts 모듈
%%writefile manage_contacts.py

contacts = []  # 연락처 관리하는 딕셔너리의 리스트

## 연락처 출력 함수 : print_contacts -
def print_contacts() :
    for dic in contacts:
        print(dic['이름'], dic['전화번호'])


## 연락처 추가 함수 : add_contacts -
def add_contacts() :
    name = input('이름 입력 : ')
    tel = input('전화번호 입력 : ')

    dic = {'이름':name, '전화번호':tel }
    contacts.append(dic)


## 전화번호 검색 함수 : search_contacts -
def search_contacts() :
    search_name = input("검색할 이름을 입력하시오:")
    for dic in contacts:
        print(dic)
        if search_name == dic['이름'] :
            print(dic['이름'], dic['전화번호'])
            return
    print("찾을 수 없습니다.")
```

### **pandas 활용: 타이타닉 생존자 분석**
1. pandas 모듈 가져오기
2. 타이타닉 데이터 가져오기
3. 데이터 분석
    1. 성별에 따른 생존율 분석
    2. 객실등급에 따른 생존율 분석
    3. 성별과 객실등급에 따른 생존율 분석

```python
# 1. pandas 모듈 가져오기
import pandas as pd

# 2. 타이타닉 데이터 가져오기
##    인터넷에 있는 타이타닉 데이터셋 CSV 주소
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
## 데이터 읽어오기
df = pd.read_csv(url)

# 데이터의 처음 5행 확인
display(df.head())         # display : 코랩에서 print보다 예쁘게 표 보이기
```

### **1. 성별에 따른 생존율 분석**

```python
# 성별(Sex)로 그룹을 묶고, 생존여부(Survived)의 평균(mean) 구하기
gender_survival = df.groupby('Sex')['Survived'].mean()
display(gender_survival)
```

### **2. 객실등급에 따른 생존율 분석**


```python
# 객실 등급별 생존율 평균 계산
pclass_survival = df.groupby('Pclass')['Survived'].mean()
display(pclass_survival)
```

###     **3. 성별과 객실등급에 따른 생존율 분석**


```python
# 성별 및 객실 등급별 생존율 집계
grouped = df.groupby(['Pclass', 'Sex'])['Survived'].mean().unstack()
display(grouped)
```
