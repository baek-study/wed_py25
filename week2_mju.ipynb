# **2. 자료형(Data Type)**
- 정수형, 실수형, 문자열형, 논리형

## 2.1. **정수형(int)**
* 소수점 없는 숫자, 파이썬 정수는 크기 제한 없음

```python
# 정수형 연습
print(123)
print(123_456_789)
print(1234_5678_9123)
print(100**100)
```

## 2.2. **실수형(float)**
* 소수점 포함한 수, 부동소수점(3.14*e12 <-3.14 * 10<sup>12</sup>)

```python
# 실수형 연습
print(3.14)
print(3.14e12)
print(3.14e-12)
```

## 2.3. **문자열형(str)**
* 단일 문자 집합, 작은따옴표/큰따옴표 로 감싼 문자 묶음

```python
# 문자열형 연습
print('hello')
print("world")
print('hello world')  # 문자열 내 빈칸도 ok
```

## 2.4. **논리형(bool)**
* 참과 거짓으로 표현, True(참), False(거짓)

```python
# 논리형 연습
print(True)
print(False)

# 문자열
print('True')

# 오류
print(true)
```

## 2.5. **자료형 확인 함수 type()**
* 정수(int), 실수(float), 문자열(str), 논리(bool)

```python
# 자료형 확인 연습
print(type(3.14))
print(type(1234))
print(type("hello"))
print(type(x))
```

```python
# my_int에 4, my_float에 2.5, my_string 에 "hello"를 대입
my_int=4
my_float=2.5
my_string="hello"

# 각 변수의 자료형을 확인 하시오
print(type(my_int))
print(type(my_float))
print(type(my_string))
```
