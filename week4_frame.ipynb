1분은 60초, 1시간은 3600초  힌트(몫과 나머지를 여러번 이용)
---
변환하고자 하는 시간(초)? (사용자 입력) <br>
000 초는 00 시간 00 분 00 초이다.


```python
# 사용자에게 초 단위의 시간을 입력받안 시-분-초 형태로 출력
SEC_PER_HOUR = 3600
SEC_PER_MINUTE = 60

total_seconds = int(input('변환하고자 하는 시간(초)?'))

hours = total_seconds // SEC_PER_HOUR
remaining = total_seconds % SEC_PER_HOUR

minutes = remaining // SEC_PER_MINUTE
seconds = remaining % SEC_PER_MINUTE

print(total_seconds, "초는", hours, " 시간",  minutes, " 분",   seconds, " 초이다.")
```

### **1) 인덱싱 : 문자열[index]**
- 특정위치의 개별 문자 반환

```python
## 1. 인덱싱 : [index] 특정위치 반환 - 0 부터 시작
msg = '안녕하세요'

# msg의 0 번째 인덱스 출력
msg[0]

# msg의 -1번째 인덱스 출력
msg[-1]

# msg에서 '하' 를 출력 - 양의 인덱스, 음의 인덱스 이용
msg[2]
msg[-3]
```

### **2) 슬라이싱 : 문자열[시작:종료]**
* **기본** : 문자열[시작:종료] - 시작부터 종료 직전(종료-1) 까지 부분 문자열 반환
* **시작 생략** : 문자열[:종료] -  첫문자 부터
* **종료 생락** : 문자열[시작:] -  마지막 문자 까지

```python
## 2) 슬라이싱1 : [시작:종료] 시작부터 끝 직전(끝-1) 위치까지 부분 문자열 반환
msg = '안녕하세요'

# 문자열 '하세'  출력
msg[2:4]
# 문자열 '녕하세' 출력
msg[1:4]
# 문자열 '하세요' 출력
print(msg[2:5])
# 문자열 '안녕' 출력
print(msg[0:2:1])

# 문자열 시작 생략 가능 : 문자열 첫 문자부터
print(msg[:2:1])

# 문자열 끝 생략 : 문자열 마지막 문자까지
print(msg[2::1])
print(msg[::1])
print(msg[::2])
print(msg[::3])
print(msg[::4])
print(msg[::-1])
```

```python
## 2) 슬라이싱2 : 문자열을 2부분으로 나누기

msg = "20250425Rainy"

# date
date = msg[:8]
# weather
weather = msg[8:]

# year
year = date[0:4]
# month
month = date[4:6]
# day
day = date[6:]

print("year : ", year, "month : ", month, "day : ", day, "weather : ", weather)
```

## **도전 실습 과제**

자신이 좋아 하는 노래를 이용하여 아래 문자열을 사용자에게 입력 받은 후,  문자열을 이용하여 연산( +, *, in )과 문자열 관련 함수 중 적어도 3개를 실습 해보기 <br>
*   출력은 f 문자열 서식 사용

* 문자열1 : 자신이 좋아하는 노래제목<br>
* 문자열2 : 가수<br>
* 문자열3 : 노래내 주요 단어<br>
* 문자열4 : 노래 가사<br>
-
* 문자열 이름에 자신의 학번 뒤 1자리 넣기<br>
* 변수 이름을 명확하고 창의적으로 짓기<br>
Ex) title, song_title, song_name
