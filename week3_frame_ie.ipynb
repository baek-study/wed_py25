# **2. 할당(대입)연산자**
* 우측 표현식의 결과값을 좌측 변수에 대입

## **2.1 단순 할당 - 하나의 변수에 하나의 값**
자료 - 단순 값 또는 다른 변수값 <br>
수식 - 값/값, 변수/변수, 값/변수 등 수식 결과값 대입 <br>
함수 - 함수 결과 값 대입

```python
# 단순값 대입 - n1에 1 넣기
n1 = 1

# 다른 변수값 대입 - n2에 n1 값 넣기
n2 = n1

# 값-값 연산 : n1 변수에 1과 2를 더한 값 넣기
n1 = 1+2

# 변수-값 연산 : n2 변수에 n1에 4를 나눈 값 넣기
n2 = n1/4

# 변수-변수 연산 : max_value 에 max(n1, n2) 함수 결과 값 넣기
max_value = max(n1, n2)
```

### **2.2 다중할당문**
- 한 값에 여러 변수 변수 초기화 <br>
- 여러 변수에 각각 값 변수 초기화 <br>
- 두 변수 맞교환

```python
# 다중 할당문1 - 한개의 값을 여러 변수에 할당
#  x, y, z에 모두 1을 할당
x=y=z=1
print("x=", x,"y=", y,"z=", z)

z=12
y=z
x=y
print("x=", x,"y=", y,"z=", z)

# 다중 할당문2 - 여러변수에 각각의 값 할당
# x에는 1, y에는 2 넣기
x = 1
y = 2

x, y = 1, 2
print("x=", x,", y=", y)

# x,y,z에 각각 10, 20, 30 넣기
x, y, z = 10, 20, 30
print("x=", x,"y=", y,"z=", z)

# 다중 할당문 3 - 두 변수 맞교환
# x, y의 값을 서로 바꾸기 - 파이썬의 독특한 방법
print("x=", x,", y=", y)
x, y = y, x
print("x=", x,", y=", y)


# 다른 언어에서 맞교환 - temp 이용
temp  = x
x = y
y = temp

print("x=", x,", y=", y)

#num1, num2, num3 = 1,2  #오류

# num1, num2 = 1         #오류

num1 = 1, 2, 3, 4        # 특수 경우 - 군집 데이터(튜플)
print(num1)

# num1 = 100+200; num2 = int("202")
num1, num2 = 100+200, int("202")
print(num1)
print(num2)
```

## **2.4 복합대입연산자**
+=처럼 대입 연산자와 다른 연산자를 합쳐 놓은 연산자 <br>

```python
num = 10

# 복합대입 num = num + 2
num += 2

# 복합대입 num = num - 2
num -= 2

# 복합대입 num = num * 2
num *= 2

# 복합대입 num = num / 2
num /= 2
```

# **3부 관계 연산자**

## **3.1 관계 연산자**
* 두 개의 자료 또는 값(피연산자)을 비교하는 연산자<br>
* 결과값은 항상 참(true) 아니면 거짓(false)

```python
# 관계 연산자 실습 1
# n1에 10, n2에 100 저장
n1 = 10
n2 = 100

# n1과 n2가 같은지
print(n1 == n2)

# n1과 n2가 같지 않은지
print( n1 != n2)

# n1과 n2보다 작은지
print(n1 < n2)

# n1과 n2보다 큰지
print(n1 > n2)

# n1과 n2보다 작거나 같은지
print(n1<=n2)

# n1과 n2보다 크거나 같은지
print(n1>=n2)
```

```python
# 관계 연산자 실습 2
# radius에 100 할당
radius=100

# flag1 에 radius가 32 보다 큰지 확인
flag1 = (radius>32)

# flag2 에 radius가 32 와 같은지 확인
flag2 = (radius==32)

# flag3 에 radius가 32 와 같지 않으지 확인
flag3 = (radius!=32)

# flag 출력
print("radius>32", flag1, "radius==32", flag2, "radius!=32", flag3)
```

## **3.2 관계/비교 연산자 주의 사항** ###


1.   == : 같다, = : 대입 연산자
2.   논리형 비교 : ==, !=
3. 실수시 비교 : 근사값 저장됨



```python
# 1. == : 같다, = : 대입 연산자
n1 = 3
n2 = 3
# 같다는 표현
print(n1 == n2)
# 대입시는 ?
```

```python
# 2. 실수시 비교 : 근사값 저장됨

num = 1.2-1.0
print(num == 0.2)

# 실수 비교는 오차 범위 내로
is_equal = ( (0.2 - num) < 0.0001 )
print(is_equal)
```
