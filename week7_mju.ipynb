## **2.1 중첩 if : if 안에 if**

 if 조건평가식 : <br>
 [ ]if 조건평가식 : <br>
 [ ][빈칸] 문장 <br>
 [ ]else : <br>
 [ ][빈칸] 문장 <br>
else : <br>
[빈칸] 문장 <br>

만약 배송지가 한국이면
- 상품의 가격이 2만원 이상이면 배송비는 없고
그렇지 않으면 3000원의 배송비가 붙는다.

그렇지 않으면
- 8000원의 배송비가 붙는다.<br>
image.png


```python
# 중첩 if 예제1
country = 'usa'
cost = 25000

# 외부 if : 만약 배송지가 한국 이면
#           그렇지 않으면 shipping_cost가 8000
if country == 'korea':
    print(f'korea : {country}')
    # 내부 if : 만약 가격이 20000만원 이상 이면 shpping_cost 가 0,
    #           그렇지 않으면 shipping_cost가 3000
    if cost >= 20000 :
        shipping_cost = 0
    else :
        shipping_cost = 3000
else :
    print(f'not korea : {country}')
    shipping_cost  = 8000


print(f'나라는 {country}, 가격은 {cost}, 배송료는 {shipping_cost}')
```

## **[예제 플러스]**

- 계산기 프로그램를 작성하시오.  <br>
[1. 사용자 입력]
- 사용자에게 두 정수 입력 (num1, num2)
- 사용자에게 연산자(+, -, *, /)
 입력 (op) <br>
[2. 연속 if]
- 만약 op 가 + 이면 result 에 더하기 연산수행
- 그렇지 않고 - 이면result 에 빼기 연산수행
- 그렇지 않고 * 이면result 에 곱하기 연산수행
- 그렇지 않고 / 이면result 에 나누기 연산수행
- 그렇지 않으면 "무효입니다" 출력

```python
# 연속 if 예제 2
# 1.  정수 입력 2개 - num1, num2
num1 = int(input('첫번째 정수를 입력하시오'))
num2 = int(input('두번째 정수를 입력하시오'))

# 2. 연산자 입력 - op
op = input('사칙연산자를 입력하시오')

# 3. 연속 if  + , -, *, /    result = num1 + num2
if op == '+' :
    result = num1 + num2
    print(f'계산결과는{result = }입니다')
elif op =='-' :
    result =num1 - num2
    print(f'계산결과는{result = }입니다')
elif op =='*' :
    result =num1 * num2
    print(f'계산결과는{result = }입니다')
elif op =='/' :
    result =num1 / num2
    print(f'계산결과는{result = }입니다')
else :
    print('무효입니다')
```

## **[예제2]**
 가위, 바위, 보 게임을 통해서 컴퓨터와 대결해 봅시다.
* 잘 알겠지만, 가위바위보 게임은 다음의 규칙을 따릅니다. 나를 기준으로 컴퓨터가 내는 것으로 확인하면 됩니다.

```python
# my_hand vs com_hand

import random

# 사용자의 가위/바위/보 중 하나를 입력받아 myHand 변수에 저장
my_hand = input('가위/바위/보 중 하나를 입력하시오')

# 컴퓨터의 가위/바위/보는 랜덤하게 추출하여 comHand에 저장
com_hand = random.choice(['가위','바위','보'])

print(f'{my_hand=}, {com_hand=}')

# 게임 진행하기 위해 사용자의 가위/바위/보에 따라 if문을 사용함
# my_hand가 가위 이면
# 그렇지 않고 my_hand가 바위 이면
# 그렇지 않고 my_hand가 보 이면
# 그렇지 않으면 - '무효입니다'

if my_hand == '가위':
    if com_hand == '가위':
        print("비김")
    elif com_hand == '바위':
        print("졌습니다")
    else :
        print("이겼습니다")
elif my_hand == '바위':
    if com_hand == '가위':
        print("이겼습니다")
    elif com_hand == '바위':
        print("비김")
    else :
        print("졌습니다")
elif my_hand == '보':
    if com_hand == '가위':
        print("졌습니다")
    elif com_hand == '바위':
        print("이겼습니다")
    else :
        print("비김")
else :
    print("무효 입니다")
```
