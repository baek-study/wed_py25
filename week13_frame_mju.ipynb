## **심화 예제 : calc_arithmetic.py**
- 두 수의 합, 차, 곱, 나누기 함수  

```python
# 모듈 calc_arithmetic 만들기
%%writefile calc_arithmetic.py

# add : 매개변수 num1, num2,  리턴 : 합
def add(num1, num2) :
    result = num1 + num2
    return result

# subtract : 매개변수 num1, num2,  리턴 : 차
def subtract(num1, num2) :
    result = num1 - num2
    return result

# multiply : : 매개변수 num1, num2,  리턴 : 곱
def multiply(num1, num2) :
    result = num1 * num2
    return result

# divide : : 매개변수 num1, num2,  리턴 : 나누기
def divide(num1, num2) :
    if num2 != 0 :
        result = num1/num2
    else:
        result = None
    return result
```

# **1부 딕셔너리(dictionary) 기본 구조**


## **1.1 딕셔너리 개요**
* 키와 값의 쌍으로 구성,
* 키는 유일
* 순서 없음

## **1.2 딕셔너리 생성/요소접근**

* 생성 : dic =  {키1:값1, 키2:값2,... }
* 요소접근 : dic[키]

```python
# 예제 - 딕셔너리 생성1 - dics - 빈딕셔너리
dics = { }

# 예제 - 딕셔너리 생성2 - capitals -  "korea":"seoul", "USA":"washington", "UK":"london"
capitals = {'korea':'seoul', 'usa':'washington', 'uk':'london'   }

# capitals 출력 & capitals 자료형 출력
print(f'{capitals=}')
print(type(capitals))

# 예제 - 딕셔너리 생성3 - d - 'python':'파이썬', 'basic':'기초', 'programming':'프로그래밍'
d = {'python':'파이썬', 'basic':'기초', 'programming':'프로그래밍'}

# d 출력 & d 자료형 출력
print(f'{d=}')
print(type(d))

# 예제 - 딕셔너리 요소 접근   딕셔너리이름[키]
## 딕셔너리 d에서 'python'키에 연결된 값 출력.
print("python 키의 값 : ", d['python'])
## 딕셔너리 d에서 'basic'키에 연결된 값 출력.
print(d['basic'])
## 딕셔너리 d에서'c++'키에 연결된 값 출력 - 키에 없는 값 출력하는경우
print(d['c++'])
## 딕셔너리 d에서 리스트 처럼 인덱스 0 쓰면? - 키가 0 이 있는지?
```

## **1.3 요소 추가/수정**
* 추가 : dic[새로운 키] = 값
* 수정 : dic[기존 키] = 새로운 값

```python
d = {'python':'파이썬', 'basic':'기초', 'programming':'프로그래밍'}

# 예제 - 딕셔너리 요소 추가
## 딕셔너리 d에 새로운 'java'키에 '자바'값 연결 -> d 출력
print(f'추가 전 : {d=}')
d['java'] = '자바'
print(f'추가 후 : {d=}')

## 딕셔너리 d에 새로운 'c++'키에 '씨뿔뿔'값 연결 -> d 출력
d['c++'] = '씨뿔뿔'
print(f'추가 후 : {d=}')

# 예제 - 딕셔너리 요소 수정
## 딕셔너리 d에 기존 'python'키에 새로운값 'PYTHON' 연결 -> d 출력
d['python']='PYTHON'
print(f'python 수정 후 : {d=}')
## 딕셔너리 d에 기존 'java'키에 새로운값 '커피' 연결 -> d 출력
d['java']='커피'
print(f'java 수정 후 : {d=}')
```

## **1.4 유효성 검사**

* 유효성 검사 : 키 in dic


```python
d = {'python':'파이썬', 'basic':'기초', 'programming':'프로그래밍'}
# 예제 - 딕셔너리 유효성 검사
## 딕셔너리 d에 키 'python'이 있는지  결과 출력
print(f"{'python' in d =} ")
## 딕셔너리 d에 키 'python'이 없는지  결과 출력
print('python' not in d)

## key 변수에 0 값 할당
key = 'python'

# if~ else :
## key가 딕셔너리 d에 있다면 key에 대한 value값 출력 글
## 그렇지 않으면 '오류:유효하지 않은 키' 출력
if key in d :
    print(f"{key}의 값", d[key])
else :
    print('오류:유효하지 않은 키')
```

# **2부. 딕셔너리 메소드와 반복**

## **2.1 딕셔너리 메소드 - 접근/삭제**
- 요소 접근 : d.get(키)
- 요소 삭제 : d.pop(키), [참고] del 딕셔너리[키]
- 요소 목록 : 키-d.keys(), 값-d.values(), 키,값-d.items()

```python
d = {'python':'파이썬', 'basic':'기초', 'programming':'프로그래밍'}
# 예제 - 딕셔너리 요소 접근 메소드
## 딕셔너리 d에서 'python'키에 연결된 값 출력.
print( d.get('python') )  # d['python']

## 딕셔너리 d에서 'basic'키에 연결된 값 출력.
print( d.get('basic') )  # d['basic']

## 딕셔너리 d에서'c++'키에 연결된 값 출력 - 키에 없는 값 출력하는경우
print( d.get('c++'))

# 예제 - 딕셔너리 요소 삭제 메소드
## 딕셔너리 d에서 'programming'키에 연관된 값 리턴 후 삭제 -> d 출력
print(f'삭제전 : {d=}')
word = d.pop('programming')
print(f'programming 삭제후 : {d=}, {word=} ')

# 예제 - 딕셔너리 요소 삭제 키워드 del - 첨자 이용 - 'basic'키와 값 삭제
del d['basic']
print(f'basic 삭제후 : {d=}')
```

## **2.2 딕셔너리 메소드 - 목록 - 뷰객체**
* 키 목록 : d.keys()
* 값 목록 : d.values()
* 키,값 목록 : d.items()

```python
d = {'python':'파이썬', 'basic':'기초', 'programming':'프로그래밍'}
# 예제

# 키 목록 출력
print(d.keys())

# 값 목록 출력
print(d.values())

# 키,값 목록 출력
print(d.items())
```

```python
# 딕셔너리와 반복문
d = {'python':'파이썬'  ,'basic':'기초'   , 'programming':'프로그래밍'}
print('\n=========================')

# 예제1. 딕셔너리 객체 자체 이용한 for - 키와 값
for key in d :
    print(key, d[key])

print('\n=========================')
# 예제2. 딕셔너리의 키 목록을 이용한 for - 키와 값
for key in d.keys() :
    print(key, d[key])

print('\n=========================')
# 예제3. 딕셔너리의 항목(키, 값) 목록을 이용한 for - 키와 값
for key, value in d.items() :
    print(key, value)
```
