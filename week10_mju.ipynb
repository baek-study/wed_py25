###**실습문제**
 -  임의의 수(1~100)를 생성 한 후, 추측한 숫자를 입력하면, 큰지 작은 지 판단해서, 정확할 때 까지 추측 하는 간단한 게임
---
1에서 100 사이의 숫자를 추측 해 보세요 : 50<br>
정답은 더 작습니다.<br>
 1에서 100 사이의 숫자를 추측 해 보세요 : 45<br>
정답은 더 큽니다.<br>
 1에서 100 사이의 숫자를 추측 해 보세요 : 47<br>
정답입니다. 3번만에 맞추셨군요!<br>
image.png


```python
# 컴퓨터가 임의의 수 선택

import random

answer = random.randint(1, 100)
count = 0
print(f'답은 {answer}')
while True :
    count += 1
    guess = int(input("1~100 사이 숫자를 추측해 보세요 : "))
    # if
    if answer == guess :
        print(f"축하합니다. 정답입니다. {count}번에 맞췄습니다.")
        break
    elif answer > guess :
        print("정답이 더 큽니다.")
    else :
        print("정답이 더 작습니다.")
```

#### **[예제1] 기호를 사각형 모양으로 출력** <br>

```python
# 예제1 - 중첩 for : 기호를 사각형 모양으로 출력

for height in range(5) :     # 0, 1, 2, 3, 4
    for width in range(4) :  # 0, 1, 2, 3
        print('*', end = "")
    print()                  # 한 줄 띄기
```

### **[예제2]  기호를 피라미드 모양으로 출력** <br>
\* <br>
\*\* <br>
\*\*\* <br>
\*\*\*\* <br>

```python
# 예제2 - 중첩 for : 기호를 피라미드 모양으로 출력 <br>
# 중첩 for

for height in range(5):
    for width in range(height+1):
        print('*',end='')
    print()

# 단일 for - f문자열
for height in range(5) :
    print('*'*(height+1))
```

### **[예제4] 구구단 예제** <br>
2\*1=2 <br>

...   <br>

```python
#예제4 - 중첩 for : 구구단

for dan in range(1, 10) :
    for gob in range(1, 10) :
        print(f'{dan} * {gob} = {dan*gob}')
```

### **[예제4] 구구단 예제 출력 변형** <br>

```python
#예제4 변형 - 중첩 for : 구구단

for gob in range(1, 10) :
    for dan in range(1, 10) :
        print(f'{dan}*{gob}={dan*gob:3d}', end = '    ')
    print()
```

### **[실습문제]** <br>
2! ~ 5! 계산 결과를 출력하는 프로그램을 작성하시오<br>
2!의 결과: 2<br>
3!의 결과: 6<br>
4!의 결과: 24<br>
5!의 결과: 120<br>



```python
#실습 - 중첩 for : 2! ~ 5! 계산 결과를 출력
for num in range(2, 6) :
    result = 1                  # 초기값
    for i in range(1, num+1) :
        result *= i

    print(f' {num}!은 { result} 입니다')
```

```python
#1 ~ num 더하기
num =4

for num in range(2, 6) :
    result = 0                  # 초기값
    for i in range(1, num+1) :
        result += i

    print(f' {num}의 합은 { result} 입니다')
```

```python
# 예제2 ... 무한 루프 + \break
# 메뉴 선택... 종료
# "메뉴를 선택 하시오 : "

while True :
    choice = int(input("메뉴를 선택 하시오 : "))

    if choice == 0 :
        print("종료합니다")
        break

for i in range(1, 100) :
    if i == 10:
        print(f"종료 : {i}")
        break
```

## **2.3 보조제어 continue**
- continue 이후 문장을 무시하고 다시 반복문 시작

```python
# 예제1 ... continue
# 100 이하 정수 중 홀수만 출력 - continue 이용

for i in range(101) :
    if i%2 == 0 :
        continue
    print(i, end = " " )

print()

for i in range(101) :
    if i%3 != 0 :
        continue
    print(i, end = " ")
```

## **2.4 보조제어 pass**

# **3부. 반복과 리스트 & 리스트 함축**

## **3.1 반복과 리스트**
*   for 변수 in 리스트 : <br>
&emsp;&emsp;;문장들

```python
# 예제1 - 반복과 리스트
# [1,2,3,4,5]의 요소를 출력

for i in [1,2,3,4,5] :
    print(i)
```

```python
nums = []

for i in [1,2,3,4,5] :
    nums.append(i)

print(nums)

nums_lc = [ i for i in [1,2,3,4,5]  ]
print(nums_lc)

nums_lc1 = [ i for i in [1,2,3,4,5] if i%2 == 0  ]
print(nums_lc1)

nums_lc2 = [ i if i%2 == 0 else 0 for i in [1,2,3,4,5]   ]
print(nums_lc2)
```

```python
nums2 = [] # 기존 요소 + 10

for i in [1,2,3,4,5]:
    nums2.append(i+10)
print(nums2)

nums2_lc = [ i+10 for i in [1,2,3,4,5] ]
print(nums2_lc)

nums2_lc1 = [ i+10 for i in [1,2,3,4,5] if i%2 == 0 ]
print(nums2_lc1)

nums2_lc2 = [ i+10 if i%2 == 0 else -1 for i in [1,2,3,4,5]  ]
print(nums2_lc2)
```

### **예제 - 리스트 요소를 제곱하는 리스트 생성**

```python
# 예제2 - 반복과 리스트
# [1,2,3,4,5] 요소를 제곱하여 squares 생성

squares = [] #  기존 요소를 제곱한 것

for i in [1, 2, 3, 4, 5]:
    squares.append(i*i)

print(squares)
```

## **3.2 리스트 함축**
- for 반복문을 간결하게 한줄로 새로운 리스트 생성


### **3.2.1 일반적인 형식**
-  [ 수식 for 변수 in 리스트]

```python
# 예제1 - 리스트 요소 제곱

squares_lc = [ i*i for i in [1, 2, 3, 4, 5] ]
print(squares_lc)
```

### **3.2.2 리스트 함축 + if 조건문 (필터링)**
-  [ 수식 for 변수 in 리스트 if 조건문]

```python
# 예제2 - 리스트 요소 중 짝수만 제곱

squares_lc1 = [ i*i for i in [1, 2, 3, 4, 5] if i%2 == 0 ]
print(squares_lc1)
```

### **3.2.3 리스트함축 + if~else조건문(값변환)**
-  [ 수식1 if 조건 else 수식2 for 변수 in 리스트]

```python
# 예제3 - 리스트 요소가 짝수면 제곱하고 홀수면 0 으로 값 변환

squares_lc2 = [ i*i if i%2 == 0 else 0   for i in [1, 2, 3, 4, 5]  ]
print(squares_lc2)
```
