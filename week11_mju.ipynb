1~100 에서 3의 배수를  triples 리스트에 넣고자 할때, 리스트 함축은?

```python
# 리스트 함축 예제
# 1. 1~100 까지의 숫자 리스트
nums = [ x for x in range(1, 101) ]
print(f'{nums=}')

# 2. 1~100 까지 중 3의 배수의 리스트
triples = [ x for x in range(1, 101)   if x%3 ==0 ]
print(f'{triples=}')

# 3. 1~100 까지 중 3의 배수인 경우 그대로, 그외는 0 으로 된 리스트
temp = [ x if x%3 ==0 else 0 for x in range(1, 101) ]
print(f'{temp=}')

#4. 1~100 까지 중 2의 배수인 경우  "even", 아닌 경우 "odd" 로 된 리스트
odd_even = ["odd"if x%2==1 else "even" for x in range(1,101)]
print(f"{odd_even=}")
```

```python
# 리스트 연습
# fruits 빈리스트
fruits=[]
# fruits에 apple, banana, cherry 추가
fruits.append('apple')
fruits.append('banana')
fruits.append('cherry')
print(f'{fruits=}')

# fruits의 인덱스 1에 pear 삽입
fruits.insert(1,'pear')
print(f'{fruits=}')

# fruits에서 banana 제거
fruits.remove('banana')
print(f'banana 제거후 {fruits=}')

# fruits에서 인덱스 1 값 제거 후 item 에 넣은 후 item 출력
item=fruits.pop(1)
print(f'인덱스1 값 제거 후 {fruits=}')
print(item)

# fruits에 cherry 가 있는지.. 있다면 cherry가 인덱스 어디에 있는지?
id= fruits.index('cherry')
print( f'cherry의 인덱스 : {id}')

# fruits에 있는 각 요소들의 길이를 length_chars 리스트로 넣을 때, 리스트 함축
length_chars = [ len(fruit) for fruit in fruits ]
print(f'{length_chars = }')

# market 이라는 빈리스트 생성
market = []
# market 에 fruits 추가 - 이중 리스트
market.append(fruits)

# market 이용해서 각각 요소 출력
print(f'{market = }')
print(f'{market[0] = }')
print(f'{market[0][0] = }, {market[0][1] = }')
```

# **실습문제 : 친구 연락처 관리**
- 친구 연락처를 관리하는 프로그램을 작성한다.   
 - 다음과 같은 메뉴를 가짐


image.png

----------
메뉴 선택 : 2 <br>
이름 입력 :  홍길동<br>
전화 번호 입력 : 111-1111<br>
홍길동 연락처가 추가 되었습니다. <br>
메뉴 선택 : 3<br>
삭제할 이름 입력 : 홍길동<br>
홍길동 연락처가 삭제 되었습니다.


```python
# 예제 - 친구 연락처 관리

contacts = []  # 연락처 관리하는 2 차원 리스트

while True :

    # 메뉴 출력
    print("--------------------")
    print("1. 친구 연락처 출력")
    print("2. 친구 연락처 추가")
    print("3. 친구 연락처 삭제")
    print("9. 종료")
    print("--------------------")

    menu = int(input("메뉴 입력 : "))

    if menu == 1 :
        for contact_list in contacts :
            print(contact_list)
            print(f'이름:{contact_list[0]} 전화번호 : {contact_list[1]}')

    elif menu == 2 :
        name = input("이름 입력 : ")
        tel = input("전화번호 입력 : ")
        contact_list  = [name, tel]
        contacts.append(contact_list)
        # contacts.append([name, tel])
    elif menu == 3 :
        found = False
        delete_name = input("삭제할 이름 입력 : ")

        for contact_list in contacts :
            if contact_list[0] == delete_name :
                found = True
                delete_contact = contact_list
                break

        if found :
            contacts.remove(delete_contact)

    elif menu == 9 :
        print("프로그램을 종료 합니다")
        break
    else :
        print("잘못된 선택 입니다")
```

 # **3부 매개변수 전달/값 반환**

 ## **3.1 매개변수 전달**
*   매개변수(parameter) : 함수 정의에서 인수를 전달받는 변수
*  인수(argument) : 함수 호출시 함수에게 전달되는 값

---
일반적으로, 매개변수 갯수와 인수 갯수 같음



```python
# 예제 1 - 원의 넓이를 구하는 함수 - 매개변수 : radius
# 함수 정의 - get_area
def get_area(radius) :
    area = 3.14 * radius**2
    return area

# 함수 호출 - get_area
```

```python
# 예제 2 - 사각형의 넓이를 구하는 함수 - 매개변수 : width, height
# 함수 정의 - get_rect_area
def get_rect_area(width,height):
    area = width * height
    return area

# 함수 호출 - get_rect_area
```

```python
# 예제 3 - 두 수의 합을 구하는 함수 -
# 매개변수 : v1, v2
# 함수 정의 - sum_para2
def sum_para2(v1, v2) :
    result =  v1 + v2
    return result

# 함수 호출 - sum_para2
value = sum_para2(10, 20)
print(f'10과 20을 더하면 {value} 입니다')
```

```python
# 예제 4 - 세 수의 합을 구하는 함수 - 매개변수 : v1, v2, v3
# 함수 정의 - sum_para3
def sum_para3(V1, V2, V3):
    result=V1+V2+V3
    return result

# 함수 호출 - sum_para3
value=sum_para3(10,20,30)
print(f'10과 20과 30을 더하면 {value}입니다.')
```

#### **매개변수에 기본값 저장**
- 매개변수 = 값 # 뒤에서 부터~

```python
# 예제 4 - 세 수의 합을 구하는 함수 - 매개변수 : v1, v2, v3 (기본값 0 )
# 함수 정의 - sum_para
def sum_para(v1, v2=0, v3=0) :
    print(f'{v1=}, {v2=}, {v3=}')
    result = v1 + v2 + v3
    return result

# 함수 호출 - sum_para
value1 = sum_para(10, 20, 30)
print(f'{value1=}')

value2 = sum_para(10, 20)
print(f'{value2=}')

value3 = sum_para(10)
print(f'{value3=}')
```

 ## **3.2 값 반환**
*   반환 값(return value) : 함수 수행의 결과로 호출자에게 돌려주는 값
* 결과변수 : 함수 결과를 저장하기 위한 변수
-----
* 반환값이 없는 함수
* 반환값이 1개 있는 함수
* 반환값이 2개 있는 함수

```python
# 값반환 예제1 - 반환값이 없는 함수
# 함수 정의
# 함수 이름: func1
# 매개변수 : 없음
# 반환값 : 결과(return) 없음
# 하는 일 : 없음
def func1() :
    print("func1은 리턴이 없습니다")

# 함수 호출
func1()
```

```python
# 값반환 예제2 - 반환값이 한개 있는 함수
# 함수 정의
# 함수 이름: func2
# 매개변수 : 없음
# 반환값 : result 반환
# 하는 일 : result에 100 넣음
def func2() :
    result = 100
    return result

# 함수 호출 - 결과변수(value)
value = func2()
print(f'{value=}')
```

```python
# 값반환 예제2 - 반환값이 두개 있는 함수
# 함수 정의
# 함수 이름: func3
# 매개변수 : 없음
# 반환값 : result1, result2 반환
# 하는 일 : result1에 100, result2에 200 넣음
def func3() :
    result1 = 100
    result2 = 200
    return result1, result2

# 함수 호출 - 결과변수(value1, value2)
value1, value2 = func3()

print(f"{value1 = }, {value2 = }")

value = func3()
print(f"{value=}")
```

- 양수를 입력하였을 경우 “양수 입력!”
- 음수를 입력하였을 경우 “음수 입력!”,

#### 실행 결과
정수를 입력하시오: -7 <br>
음수 입력!

```python
# decide_num   - 매개변수 number    반환값, 0, 양수, 음수

def decide_num(number) :
    if number == 0 :
        result = "0"
    elif number >0 :
        result = "양수"
    else :
        result = "음수"
    return result

# 함수 호출
number = int(input('정수를 입력하시오 : '))
value = decide_num(number)
print(f'{value = }')
```
